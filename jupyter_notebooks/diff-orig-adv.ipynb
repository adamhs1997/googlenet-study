# Image Diffs

Diff adversarial image with its original. Essentially "adv-orig".

```python
import PIL
import PIL.Image
import numpy as np
from os import listdir
from skimage.io import imsave
```

```python
# For every adv image
    # Open the image
    # Open its corresponding original
    # Perform the diff
    # Save the diff
    
# Get list of original images from each dir
img_0_dir = listdir("/home/w266ajh/Documents/top5gen/data_imageparts/")

# Get the adversarial images
adv_images = ["/home/w266ajh/Documents/jupyter_notebooks/adv_images/",
              "/home/w266ajh/Documents/jupyter_notebooks/sim_adv_images/"]
    
for folder in adv_images:
    for img in listdir(folder):
        if img == ".directory": continue
        base_image = img.split('_')[1]

        # Deterimine where the original for this adv is
        if base_image in img_0_dir:
            base_path = "/home/w266ajh/Documents/top5gen/data_imageparts/"
        else:
            base_path = "/home/w266ajh/Documents/top5gen/data_imageparts_1/"

        # Open the base image
        base = PIL.Image.open(base_path + base_image)

        # Open the adversarial
        adv = PIL.Image.open(folder + img)

        # Resize base to be 224x224 (as the adv already was)
        big_dim = max(base.width, base.height)
        wide = base.width > base.height
        new_w = 224 if not wide else int(base.width * 224 / base.height)
        new_h = 224 if wide else int(base.height * 224 / base.width)
        base = base.resize((new_w, new_h)).crop((0, 0, 224, 224))

        # Set up images as Numpy array
        base = (np.asarray(base) / 255.0).astype(np.float32)
        adv = (np.asarray(adv) / 255.0).astype(np.float32)

        # Perform the subtraction
        diff = adv - base
        diff += 0.3

        # Save the diff
        imsave("/home/w266ajh/Documents/jupyter_notebooks/image_diffs/" + img + "_diff", diff, plugin='pil', format_str='png')
```
